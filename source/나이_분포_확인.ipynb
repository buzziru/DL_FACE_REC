```python
import h5py
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
with h5py.File('/content/drive/MyDrive/DL_DATA/DL_Face_REC/data_age_train.hdf5', 'r') as file:
    keys = list(file['age'].keys())
    count = len(keys)
    ages = []
    for key in file['age'].keys():
        age = file['age'][key][()]
        ages.append(age)

ages = np.array(ages)
mean_age = np.mean(ages)
std_age = np.std(ages)
min_age = np.min(ages)
max_age = np.max(ages)
```

```python
print(mean_age)
print(std_age)
print(min_age)
print(max_age)
```

```python
plt.figure(figsize=(10, 8))
sns.histplot(data=ages, binwidth=2)
plt.xlabel('Age', fontsize=15)
plt.ylabel('Count', fontsize=15)
plt.title('Distribution of Age', fontsize=20)
plt.show()
```
