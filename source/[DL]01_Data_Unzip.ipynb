```python
from glob import glob

source_path = '/content/drive/Othercomputers/집/dataset_face/118.안면 인식 에이징(aging) 이미지 데이터/01-1.정식개방데이터/Training/01.원천데이터'
image_zips = sorted(glob(source_path + '/*.zip'))
image_zips
```

```python
label_path = '/content/drive/Othercomputers/집/dataset_face/118.안면 인식 에이징(aging) 이미지 데이터/01-1.정식개방데이터/Training/02.라벨링데이터'
label_zips = sorted(glob(label_path + '/*.zip'))
label_zips
```

```python
import zipfile

class ZipDataset():
    def __init__(self, zip_paths, dst_path):
        self.zip_paths = zip_paths
        self.dst_path = dst_path

    def extract(self):
        for path in self.zip_paths:
            with zipfile.ZipFile(path, 'r') as zip_ref:
                zip_ref.extractall(self.dst_path)
            print(path)

dst_path_image = '/content/drive/MyDrive/DL_DATA/DL_Face_REC/train/source'
zip_image = ZipDataset(image_zips, dst_path_image)
zip_image.extract()
```

```python
dst_path_label = '/content/drive/MyDrive/DL_DATA/DL_Face_REC/train/label'
zip_label = ZipDataset(label_zips, dst_path_label)
zip_label.extract()
```
